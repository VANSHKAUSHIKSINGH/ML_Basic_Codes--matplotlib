# Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
classes = ["Python","R","Machine Learning","Artifical Intelligence","Data Science"]
Class1_students = [45,15,35,25,30]
plt.pie([1])
plt.show()
```

```python
plt.pie(Class1_students)
plt.show()
```

```python
plt.pie(Class1_students, labels = classes)
plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artifical Intelligence","Data Science"]
class1_students = [45,15,35,25,30]
explode = [0.03,0,0.1,0.0]
colors = ['c','b','r','y','g']
plt.pie(class1_students, labels = classes, colors = colors)
plt.show()
```

```python
class1_students = [45,15,35,25,30]
explode = [0.03,0,0.1,0,0]
colors = ['c','b','r','y','g']
plt.pie(class1_students, labels = classes,explode = explode,colors = colors,autopct = "%0.2f%%",shadow = True,radius = 1.4,
       startangle = 270)
plt.show()
```

```python
textprops = {"fontsize":15}
plt.pie(class1_students,labels = classes,explode = explode,colors = colors,autopct = "%0.2f%%",shadow = True,radius = 1.4,
       startangle = 270,textprops = textprops)
plt.show()
```
