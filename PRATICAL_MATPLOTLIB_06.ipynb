# Ploting Bar Chart (cont.)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
Classes = ["python","R","AI","ML","DS"]
Class1_students = [30,10,20,25,10]#out of 100 student ijn each class
Class2_students = [40,5,20,20,10]
Class3_students = [35,5,30,15,15]
```

```python
style.use("ggplot")
plt.figure(figsize = (16,9))
plt.bar(Classes, Class1_students, width = 0.6,align = "edge", color = "k",edgecolor = "m",linewidth = 5,alpha = 0.9,
       linestyle = "--",label = "Classes1_students")
```

```python
plt.figure(figsize = (16,9))
plt.bar(Classes, Class1_students, align = "edge", color = "k",edgecolor = "m",linewidth = 5,alpha = 0.9,
       linestyle = "--",label = "Classes1_students")
```

```python
plt.figure(figsize = (16,9))
plt.bar(Classes, Class1_students, width = 0.6,align = "edge", color = "k",edgecolor = "m",linewidth = 5,alpha = 0.9,
       linestyle = "--",label = "Classes1_students")
plt.title("Bar Chart Of IAIP Class")
plt.xlabel("Classes")
plt.ylabel("No. of students")
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.bar(Classes, Class1_students, width = 0.6,align = "edge", color = "k",edgecolor = "m",linewidth = 5,alpha = 0.9,
       linestyle = "--",label = "Classes1_students")
plt.title("Bar Chart Of IAIP Class", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.bar(Classes, Class1_students, width = 0.2,color = "b",label = "Classes1_students")
plt.bar(Classes, Class2_students, width = 0.2,color = "g",label = "Classes2_students")
plt.title("Bar Chart Of IAIP Class", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (16,9))
classes_index = np.arange(len(Classes))
width = 0.2
plt.bar(classes_index, Class1_students, width, color = "b", label = "Classes1_students")
plt.bar(classes_index, Class2_students, width,  color = "g", label = "Classes2_students")
plt.title("Bar Chart Of IAIP Class", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (16,9))
classes_index = np.arange(len(Classes))
width = 0.2
plt.bar(classes_index, Class1_students, width, color = "b", label = "Classes1_students")
plt.bar(classes_index+width, Class2_students, width,  color = "g", label = "Classes2_students")
# plt.xtricks(Classes_index +width, Classes,rotation = 20)
plt.title("Bar Chart Of IAIP Class", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (16,9))
classes_index = np.arange(len(Classes))
width = 0.2
plt.bar(classes_index, Class1_students, width, color = "b", label = "Classes1_students")
plt.bar(classes_index+width, Class2_students, width,  color = "g", label = "Classes2_students")
plt.xticks(classes_index + width, Classes)
plt.title("Bar Chart Of IAIP Class", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.show()
```

```python
plt.bar(classes_index, Class1_students, width, color = "b",label = "Classes1_students")
plt.bar(classes_index+width,Class2_students,width,color = "g",label = "Classes2_students")
plt.bar(classes_index+width+width,Class3_students,width, color = "m",label = "Classes3_students")
plt.xticks(classes_index+width,Classes, rotation = 20)
plt.title("Bar Chart of IAIP class",fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of students ", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (16,9))
classes_index = np.arange(len(Classes))
width = 0.2
plt.barh(classes_index,Class1_students,width,color = "b",label = "Classes1_students")
plt.barh(classes_index+width,Class2_students,width,color = "g",label = "Classes2_students")
plt.barh(classes_index+width+width,Class3_students,width,color = "y",label = "Classes3_students")
plt.yticks(classes_index+width, Classes, rotation = 20)
plt.title("Bar chart of IAIP Class",fontsize = 18)
plt.ylabel("Classses",fontsize = 15)
plt.xlabel("No. of students",fontsize = 15)
plt.legend()
plt.show()
```
