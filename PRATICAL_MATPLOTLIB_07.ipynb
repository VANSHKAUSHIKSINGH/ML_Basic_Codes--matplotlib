```python
x = df_google_play_store_apps["Rating"]
y = df_google_play_store_apps["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google Play store Apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r",marker = "*")
plt.title("Google play store apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```
