# Ploting Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
# Dataset of "Indian Artifical Intelligence Production (IAIP) Class"
# Classes = ["Python","R","Artifical Intelligence","Machine Learning","Data Science"]
```

```python
Classes = ["python","R","AI","ML","DS"]
Classes1_students = [30,10,20,25,10]#out of 100 student ijn each class
Classes2_students = [40,5,20,20,10]
Classes3_students = [35,5,30,15,15]
```

```python
plt.bar(Classes, Classes1_students)
```

```python
plt.barh(Classes, Classes1_students)
```

```python
plt.bar(Classes, Classes1_students, width = 0.2)
```

```python
plt.bar(Classes, Classes1_students, width = 0.2,align = "edge",color = "y")
```

```python
plt.bar(Classes, Classes1_students, width = 0.2,align = "edge",color = "y",edgecolor = "m")
```

```python
plt.bar(Classes, Classes1_students, width = 0.2,align = "edge",color = "y",edgecolor = "m",linewidth = 5,alpha = 0.5)
```

```python
plt.bar(Classes, Classes1_students, width = 0.2,align = "edge",color = "y",edgecolor = "m",linewidth = 5,alpha = 0.9,
        linestyle = "--",label = "Class1_students")
```
